## Mengambil komentar


 YouTube Data API v3 untuk mengambil komentar dari video yang ditentukan.


```python
API_KEY = "AIzaSyD_NzKH6rwXq9N297sqAN4aTp8GwW7Di2s" # API_KEY Youtube V3

youtube = build('youtube', 'v3', developerKey=API_KEY)

comments = []
next_page_token = None
max_comments = 4000
results_per_request = 100

while len(comments) < max_comments:
    request = youtube.commentThreads().list(
        part='snippet',
        videoId=video_id,
        pageToken=next_page_token,
        maxResults=results_per_request
    )
    response = request.execute()

    for item in response['items']:
        comments.append(item['snippet']['topLevelComment']['snippet'])

    next_page_token = response.get('nextPageToken')

    if not next_page_token:
        break

print(f"Collected {len(comments)} comments.")
```

### Case Folding

```python
import re

def clean_text(text):
    # Menghapus HTML tags
    text = re.sub(r'<.*?>', '', text)
    # Remove URLs
    text = re.sub(r'http\S+|www\S+|https\S+', '', text, flags=re.MULTILINE)
    # Remove special characters and symbols (keeping basic punctuation)
    text = re.sub(r'[^\w\s.,!?;:]', '', text)
    # Convert to lowercase (Case Folding)
    text = text.lower()

    text = text.strip()
    return text


df_comments['textOriginal'] = df_comments['textOriginal'].apply(clean_text)

display(df_comments.head())
```

## Membersihkan teks komentar

### Subtask:
Membersihkan kolom teks komentar dari karakter yang tidak diinginkan, seperti tag HTML, URL, atau karakter khusus lainnya.


```python
import re

def clean_text(text):
    # Remove HTML tags
    text = re.sub(r'<.*?>', '', text)
    # Remove URLs
    text = re.sub(r'http\S+|www\S+|https\S+', '', text, flags=re.MULTILINE)
    # Remove special characters and symbols (keeping basic punctuation)
    text = re.sub(r'[^\w\s.,!?;:]', '', text)
    # Convert to lowercase
    text = text.lower()
    # Remove leading and trailing spaces
    text = text.strip()
    return text

df_comments['textOriginal'] = df_comments['textOriginal'].apply(clean_text)

display(df_comments.head())
```
