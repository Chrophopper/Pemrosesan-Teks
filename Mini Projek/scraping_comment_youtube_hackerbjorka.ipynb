## Menyimpan data

### Case Folding

Melakukan Case Folding untuk merubah kata-kata yang sama tetapi memiliki format huruf yang berbeda dengan cara mengubah menjadi huruf kecil (lowecase).

**Catatan :** Data yang telah dirubah pada parameter adalah textOriginal.

```python
import re

def clean_text(text):
    # Menghapus HTML tags
    text = re.sub(r'<.*?>', '', text)
    # Remove URLs
    text = re.sub(r'http\S+|www\S+|https\S+', '', text, flags=re.MULTILINE)
    # Menghapus Tanda Baca yang tidak dibutuhkan
    text = re.sub(r'[^\w\s.,!?;:]', '', text)
    # Konversi menjadi Case Folding
    text = text.lower()

    text = text.strip()
    return text


df_comments['textOriginal'] = df_comments['textOriginal'].apply(clean_text)

display(df_comments.head())
```
