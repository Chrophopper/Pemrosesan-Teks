## **Pra-Pemrosesan Teks**

### **Text Cleaning**
menghapus Punctuation, Emojis, and anomali simbol lainya

```python
def clean_text(text):
    # Konversi input text (default) diubah menjadi tipe data string.
    text = str(text)
    # menghapus tautan URL/menampilkan isi yang lebih penting
    text = re.sub(r'http\S+', '', text)
    # menghapus tanda mentions (@)
    text = re.sub(r'@\w+', '', text)
    # menghapus hashtags (#)
    text = re.sub(r'#\w+', '', text)
    # menghapus seluruh punctuation
    text = re.sub(r'[^\w\s]', '', text)
    # Remove extra whitespace
    text = re.sub(r'\s+', ' ', text).strip()
    return text

df_comments['text_cleaned'] = df_comments['textOriginal'].apply(clean_text)

display(df_comments.head())
```

```python
df_cleaned = df_comments[['text_cleaned']] # menampilkan parameter baru : text_cleaned
df_cleaned
```

### **Tokenisasi**

kode ini mengambil teks mentah dari setiap komentar, memecahnya menjadi daftar kata-kata individual (token), dan menyimpan daftar token ini dalam kolom baru di DataFrame

```python
df_comments['text_tokens'] = df_comments['text_cleaned'].apply(word_tokenize)
display(df_comments.head())
```

```python
df_tokens = df_comments[['text_tokens']] # menampilkan parameter tokenisasi
df_tokens
```

### **Stemming**

proses untuk mengurangi kata-kata menjadi bentuk akarnya

```python
stemmer = PorterStemmer() # algoritma stemming yang umum digunakan.

def stem_tokens(tokens): #menerapkan fungsi stem_tokens ke setiap elemen (setiap daftar token) dalam kolom text_tokens di DataFrame df_comments.
    return [stemmer.stem(token) for token in tokens]

df_comments['text_stemmed'] = df_comments['text_tokens'].apply(stem_tokens) #Hasil stemming kemudian disimpan dalam kolom baru bernama text_stemmed.
display(df_comments.head())
```

```python
df_stemmed = df_comments[['text_stemmed']] #menampilkan isi dari parameter text_stemmed yang telah diproses untuk mengurangi kata-kata menjadi bentuk akarnya
df_stemmed
```
