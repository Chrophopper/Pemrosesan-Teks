```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
path = pd.read_csv('processed_comments.csv')
path
```

```python
def remove_duplicate_words(text):
    words = text.split()
    unique_words = []
    for word in words:
        if word not in unique_words:
            unique_words.append(word)
    return ' '.join(unique_words)
```

```python
for col in path.columns:
    path[col] = path[col].apply(remove_duplicate_words)

display(path)
```

```python
display(path)
```

```python
def generate_ngrams(text, n):
    words = text.split()
    ngrams = zip(*[words[i:] for i in range(n)])
    return ' '.join(['_'.join(ngram) for ngram in ngrams])
```

```python
path['text_bigrams'] = path['text_cleaned'].apply(lambda x: generate_ngrams(x, 2))
path['text_trigrams'] = path['text_cleaned'].apply(lambda x: generate_ngrams(x, 3))
display(path.head())
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

all_text = ' '.join(path['text_cleaned'])

wordcloud = WordCloud(width=800, height=400, background_color='white').generate(all_text)

plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```
